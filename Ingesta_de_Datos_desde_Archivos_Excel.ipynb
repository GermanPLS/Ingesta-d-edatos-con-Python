```python
# Importamos librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importacion de un excel!
# leemos el archivo de excel y lo guardamos el contenido del archivo excel en un DataFrame llamado data_excel

data_excel = pd.read_excel("/content/data.xlsx")

# mostramos el contenido del DataFrame

data_excel
```

```python
# Ahora vamos a explorar y analizar los datos cargados desde el archivo de Excel

# Ver las primeras filas del dataframe
print("Primeras filas del dataframe:")
print(data_excel.head())
print()
# InformaciÃ³n general del dataframe
print("\nInformaciÃ³n general sobre el dataframe:")
print(data_excel.info())
print()
# EstadÃ­sticas descriptivas
print("\nEstadÃ­sticas descriptivas:")
print(data_excel.describe())

# Comprobar valores nulos
print("\nValores nulos por columna:")
print(data_excel.isnull().sum())
```

```python
# Vamos a analizar y explorar los datos de forma visual, utilizando grÃ¡ficos para entender mejor la distribuciÃ³n, correlaciones y posibles valores atÃ­picos en el DataFrame.

# Ver distribuciÃ³n de las variables numÃ©ricas
print("\nDistribuciÃ³n de las variables numÃ©ricas:")
data_excel.hist(bins=20, figsize=(14,10))
plt.suptitle("DistribuciÃ³n de Variables NumÃ©ricas")
plt.show()

# CorrelaciÃ³n entre variables numÃ©ricas usando numeric_only
print("\nMatriz de correlaciÃ³n:")
corr = data_excel.corr(numeric_only=True)  # O usar select_dtypes para filtrar columnas numÃ©ricas
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title("Mapa de Calor de CorrelaciÃ³n")
plt.show()

# Ver la cantidad de valores Ãºnicos por columna
print("\nValores Ãºnicos por columna:")
print(data_excel.nunique())

# Graficar la distribuciÃ³n de una variable especÃ­fica
# (por ejemplo, si hay una columna llamada 'edad', para ver la distribuciÃ³n de la edad)
if 'edad' in data_excel.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data_excel['edad'], kde=True, color='blue')
    plt.title('DistribuciÃ³n de la Edad')
    plt.xlabel('Edad')
    plt.ylabel('Frecuencia')
    plt.show()

# VerificaciÃ³n de outliers (valores atÃ­picos) usando boxplot
# Si existe una columna llamada 'ingresos', para ver los outliers en ella
if 'ingresos' in data_excel.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data_excel['ingresos'])
    plt.title('Outliers en Ingresos')
    plt.show()
```

# Comparativa entre CSV y Excel

Los formatos **CSV** (Comma-Separated Values) y **Excel** son ampliamente utilizados para almacenar y gestionar datos, pero presentan diferencias significativas en tÃ©rminos de funcionalidad, caracterÃ­sticas y usos. A continuaciÃ³n, se realiza una comparaciÃ³n detallada de ambos:

## 1. Estructura y Formato
- **CSV**:
  - Es un archivo de texto plano donde los valores estÃ¡n separados por comas (u otros delimitadores como tabulaciones o punto y coma).
  - Cada lÃ­nea representa un registro, y cada campo dentro del registro estÃ¡ separado por un delimitador.
  - No almacena formatos, fÃ³rmulas, ni imÃ¡genes.
  
- **Excel**:
  - Es un formato propietario de Microsoft (XLS, XLSX) que permite almacenar datos en hojas de cÃ¡lculo con filas y columnas.
  - Soporta fÃ³rmulas, formatos de celda (colores, estilos de texto, etc.), grÃ¡ficos, tablas dinÃ¡micas y otros elementos avanzados.
  - Es mÃ¡s adecuado para trabajos complejos, con mÃºltiples hojas y funcionalidades interactivas.

## 2. Facilidad de Uso
- **CSV**:
  - Es fÃ¡cil de leer y escribir debido a su simplicidad.
  - Los archivos CSV pueden abrirse con cualquier editor de texto o programa de hojas de cÃ¡lculo.
  
- **Excel**:
  - Requiere software especializado, como Microsoft Excel o Google Sheets, para acceder a su contenido.
  - Ofrece herramientas interactivas, como la creaciÃ³n de grÃ¡ficos y anÃ¡lisis de datos, pero estas requieren un aprendizaje adicional.

## 3. TamaÃ±o y Eficiencia
- **CSV**:
  - Los archivos CSV tienden a ser mÃ¡s ligeros, ya que no contienen datos adicionales como formatos o imÃ¡genes.
  - Son adecuados para datasets grandes cuando no es necesario mantener una estructura compleja.
  
- **Excel**:
  - Los archivos Excel pueden ser mÃ¡s grandes debido a los elementos adicionales que contienen (fÃ³rmulas, formatos, macros, etc.).
  - Sin embargo, Excel tiene lÃ­mites de filas y columnas (aunque muy amplios), lo que puede ser un inconveniente para archivos extremadamente grandes.

## 4. Compatibilidad
- **CSV**:
  - El formato CSV es muy compatible entre diferentes sistemas operativos y aplicaciones (Windows, macOS, Linux, Google Sheets, Python, R, etc.).
  - Debido a su simplicidad, los archivos CSV son fÃ¡cilmente importables y exportables desde diferentes lenguajes de programaciÃ³n y bases de datos.

- **Excel**:
  - Requiere software especÃ­fico para abrir archivos XLSX, aunque hay alternativas como Google Sheets y otras aplicaciones de hojas de cÃ¡lculo.
  - Si bien Excel es ampliamente usado, su compatibilidad con programas de cÃ³digo abierto o ciertos lenguajes de programaciÃ³n puede ser limitada en comparaciÃ³n con CSV.

## 5. Funcionalidades Avanzadas
- **CSV**:
  - No soporta funciones avanzadas como fÃ³rmulas, grÃ¡ficos o macros. Solo almacena datos.
  
- **Excel**:
  - Permite el uso de funciones avanzadas, como:
    - **FÃ³rmulas**: CÃ¡lculos automÃ¡ticos y complejos.
    - **Tablas dinÃ¡micas**: AnÃ¡lisis avanzado de datos.
    - **GrÃ¡ficos interactivos**: VisualizaciÃ³n de datos.
    - **Macros**: AutomatizaciÃ³n de tareas repetitivas.

## 6. Usos Comunes
- **CSV**:
  - Ideal para el intercambio de datos entre sistemas diferentes o para almacenar grandes volÃºmenes de datos de manera eficiente.
  - Utilizado frecuentemente en anÃ¡lisis de datos, bases de datos y programaciÃ³n.

- **Excel**:
  - Ampliamente usado en tareas que requieren anÃ¡lisis visual, como la contabilidad, presupuestos, anÃ¡lisis financiero, reportes, etc.
  - Es una herramienta ideal para quienes necesitan funcionalidades avanzadas para manipular y analizar datos de manera interactiva.

## 7. Desventajas
- **CSV**:
  - No soporta formatos ni funciones avanzadas (fÃ³rmulas, grÃ¡ficos, tablas dinÃ¡micas).
  - No es ideal para almacenar grandes cantidades de datos complejos o interrelacionados.

- **Excel**:
  - Puede ser mÃ¡s pesado y lento con grandes volÃºmenes de datos.
  - La dependencia de software propietario (Microsoft Excel) puede ser un inconveniente para algunos usuarios.
  
## Resumen Final
- El formato **CSV** es ideal cuando se necesita simplicidad, eficiencia en el tamaÃ±o del archivo, y compatibilidad entre diferentes sistemas y lenguajes.
- El formato **Excel** es mÃ¡s adecuado para tareas interactivas y anÃ¡lisis de datos complejos que requieren visualizaciÃ³n avanzada, automatizaciÃ³n y fÃ³rmulas.
