## Archivos CSV

Un archivo CSV, o **valores separados por comas**, organiza datos en filas y columnas, donde cada valor está separado por una coma.

Los Archivos CSV NO siempre tienen el mismo formato.
Algunos utilizan delimitadores diferentes o caracteres de cita:

  - Caracteres de cita: Caracteres que agrupan datos para evitar confusion.

  - Delimitadores: Diferentes caracteres que separan los valores.
